```python
def TRUE(x):
    def f(y):
        return x
    return f
```

```python
def FALSE(x):
    def f(y):
        return y
    return f
```

```python
TRUE('1')('0')
```

```python
FALSE('0')('1')
```

```python
def NOT(x):
    return x(FALSE)(TRUE)
```

```python
assert NOT(TRUE) is FALSE
```

```python
assert NOT(FALSE) is TRUE
```

```python
def AND(x):
    return lambda y: x(y)(x)
```

```python
assert AND(TRUE)(TRUE) is TRUE
```

```python
assert AND(TRUE)(FALSE) is FALSE
```

```python
assert AND(FALSE)(TRUE) is FALSE
```

```python
assert AND(FALSE)(FALSE) is FALSE
```

```python
def OR(x):
    return lambda y: x(x)(y)
```

```python
assert OR(FALSE)(FALSE) is FALSE
```

```python
assert OR(FALSE)(TRUE) is TRUE
```

```python
assert OR(TRUE)(FALSE) is TRUE
```

```python
assert OR(TRUE)(TRUE) is TRUE
```

```python
OR(TRUE)(FALSE)
```

```python
OR(FALSE)(FALSE)
```

```python
OR(TRUE)(TRUE)
```

```python
OR(FALSE)(TRUE)
```

```python
def ONE(f):
    return lambda x: f(x)
```

```python
def TWO(f):
    return lambda x: f(f(x))
```

```python
def THREE(f):
    return lambda x: f(f(f(x)))
```

```python
def incr(x):
    return x + 1
```

```python
incr(0)
```

```python
incr(incr(0))
```

```python
THREE
```

```python
THREE(incr)(0)
```
