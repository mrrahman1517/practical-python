```python
def TRUE(x):
    def f(y):
        return x
    return f
```

```python
def FALSE(x):
    def f(y):
        return y
    return f
```

```python
TRUE('1')('0')
```

```python
FALSE('0')('1')
```

```python
def NOT(x):
    return x(FALSE)(TRUE)
```

```python
assert NOT(TRUE) is FALSE
```

```python
assert NOT(FALSE) is TRUE
```
