```python
def getc(d):
    d = d.get('a')
    if d is not None:
        d = d.get('b')
        if d is not None:
            d = d.get('c')
    return d
```

```python
getc(data)
```

```python
getc({})
```

```python

def perhaps(d, func):
    if d is not None:
        return func(d)
    else:
        return None
```

```python
perhaps(data, lambda d: d.get('a'))
```

```python
perhaps({}, lambda d: d.get('a'))
```

```python
perhaps(perhaps(data, lambda d: d.get('a')), lambda d: d.get('b'))
```
